```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('seasonal_final.csv')
df.head()
```

```python
df.shape
```

```python
X=df.iloc[:,0:-1]
y=df.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1234)
X_train, X_dev, y_train, y_dev = train_test_split(X_train, y_train, test_size=0.2, random_state=1234)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

models = {
    'Logistic Regression': LogisticRegression(),
    'K-Nearest Neighbors': KNeighborsClassifier(),
    'Decision Tree': DecisionTreeClassifier(),
    'Random Forest': RandomForestClassifier(),
    'Support Vector Classifier': SVC(probability=True),  # Enable probability estimates
    'Gaussian Naive Bayes': GaussianNB()
}

for name, model in models.items():
    print(f"Training {name}...")
    model.fit(X_train, y_train)
    
    # Predict probabilities for test set
    y_pred_proba = model.predict_proba(X_dev)[:, 1]  # Probabilities of the positive class
    
    # Example evaluation (replace with your own metrics)
    print(f"\n{name} Classification Report:")
    y_pred = model.predict(X_dev)
    print(classification_report(y_dev, y_pred))
    
    # Example of other metrics (accuracy, confusion matrix)
    print(f"{name} Accuracy: {accuracy_score(y_dev, y_pred)}")
    print(f"{name} Confusion Matrix:")
    print(confusion_matrix(y_dev, y_pred))

    print("\n---------------------------------------\n")
```

```python
t=pd.read_csv('test_set_features.csv')
```

```python
t=t[['xyz_concern', 'xyz_knowledge', 'behavioral_antiviral_meds',
       'behavioral_avoidance', 'behavioral_face_mask', 'behavioral_wash_hands',
       'behavioral_outside_home', 'behavioral_touch_face',
       'chronic_med_condition', 'child_under_6_months', 'health_worker',
       'opinion_seas_vacc_effective', 'opinion_seas_risk',
       'opinion_seas_sick_from_vacc', 'age_group', 'race', 'sex',
       'household_adults', 'household_children']]
```

```python
t
```

```python
t.isnull().sum()
```

```python
for i in t.columns:
    if t[i].isnull().sum()>0:
        t[i].fillna(t[i].mode()[0],inplace=True)
```

```python
for i in t.columns:
    if t[i].dtype=='float64':
        t[i]=t[i].astype(int)
```

```python
def age_lab(i):
    if i=='65+ Years':
        return 5
    elif i=='55 - 64 Years':
        return 4
    elif i=='45 - 54 Years':
        return 3
    elif i=='18 - 34 Years':
        return 1
    else:
        return 2
def sex_lab(i):
    if i=='Female':
        return 1
    else:
        return 0
def race_lab(i):
    if i=='White':
        return 3
    elif i=='Black':
        return 2
    elif i=='Hispanic':
        return 1
    else:
        return 0
t['age_group']=t['age_group'].apply(age_lab)
t['race']=t['race'].apply(race_lab)
t['sex']=t['sex'].apply(sex_lab)
```

```python
svm=SVC(probability=True)
svm.fit(X_train, y_train)
    
# Predict probabilities for test set
y_pred_proba = svm.predict_proba(X_test)[:, 1]  # Probabilities of the positive class
    
# Example evaluation (replace with your own metrics)
print(f"\n Classification Report:")
y_pred = svm.predict(X_test)
print(classification_report(y_test, y_pred))
    
# Example of other metrics (accuracy, confusion matrix)
print(f"Accuracy: {accuracy_score(y_test, y_pred)}")
print(f"Confusion Matrix:")
print(confusion_matrix(y_test, y_pred))

print("\n---------------------------------------\n")
```
