```python
keep_col=[]
drop_col=[]
for i in X.columns:
    if (X[i].isnull().sum()/X.shape[0])*100<5:
        keep_col.append(i)
    else:
        drop_col.append(i)
keep_col
```

```python
X=X[keep_col]
X
```

```python
fig,ax=plt.subplots(figsize=(20,15))
c=X.corr(numeric_only=True).abs()
ax=sns.heatmap(c,annot=True,center=0,cmap='Blues',vmin=-1,vmax=1)
```

```python
y.columns
```

```python
mask=np.triu(np.ones_like(c,dtype=bool))
tri_df=c.mask(mask)
tri_df
```

```python
to_drop=[i for i in tri_df.columns if any(tri_df[i]>0.4)]
print(to_drop)
to_drop.append('respondent_id')
```

```python
X=X.drop(to_drop,axis=1)
```

```python
fig,ax=plt.subplots(figsize=(20,15))
ax=sns.heatmap(X.corr(numeric_only=True).abs(),annot=True,center=0,cmap='Blues',vmin=-1,vmax=1)
```

```python
train_seasonal=X
train_seasonal['seasonal_vaccine']=y['seasonal_vaccine']
train_xyz=X
train_xyz['xyz_vaccine']=y['xyz_vaccine']
```

```python
train_seasonal.to_csv('seasonal.csv',index=False)
train_xyz.to_csv('xyz.csv',index=False)
```
